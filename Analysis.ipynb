# Analytics


A correlation coefficient means that there is a very strong positive correlation between these two variables. It suggests that as the number of unique contributors increases, the amount of donations in USD also tends to increase. This could imply that the project is gaining popularity and attracting more supporters who are contributing more donations.

However, it is important to note that correlation does not necessarily imply causation. In other words, just because two variables are strongly correlated, it does not mean that one variable is causing the other. There could be other underlying factors that are driving the relationship between these two variables. Therefore, it is important to interpret the correlation in the context of the data and the project.

```python

df = pd.read_csv('overall.csv')
print(df.describe())
plt.hist(df['amountUSD'])
plt.xlabel('Amount Donated (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Amount Donated')
plt.show()
```

```python
with_matching = df[df['matchAmountUSD'] > 0]['amountUSD']
without_matching = df[df['matchAmountUSD'] == 0]['amountUSD']
t, p = stats.ttest_ind(with_matching, without_matching)
print(f'T-test results: t = {t}, p = {p}')

contingency_table = pd.crosstab(df['uniqueContributors']>0, df['votes']>0)
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f'Chi-square test results: chi2 = {chi2}, p = {p}, dof = {dof}')
```

```python
X = df[['votes', 'uniqueContributors']]
y = df['amountUSD']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
